## **Problem # 1.10.** 
Analyze the `Carbon_West` ([http://stat4ds.rwth-aachen.de/data/Carbon_West.dat](http://stat4ds.rwth-aachen.de/data/Carbon_West.dat)) data file at the book’s website by **(a)** constructing a frequency distribution and  a histogram, **(b)** finding the mean, median, and standard deviation. Interpret each.

**(a) Answer Below**

```python
## Download the File from the External URL using the code below

import requests
import os
import pandas as pd
import matplotlib.pyplot as plt


## Specify the File URL & Local File Name
data_url = "http://stat4ds.rwth-aachen.de/data/Carbon_West.dat"
file_name = "Carbon_West.dat"

## Function to Download the File from the URL and Preview the Data
def download_file(file_url,file_name):
    try:
        response = requests.get(file_url)
        response.raise_for_status()

        ## Write to the File locally
        with open(file_name, "wb") as f:
            f.write(response.content)
        
        ## Verify if the file got downloaded successfully or not
        if file_name in os.listdir('.'):
            print("File downloaded successfully.\n")
            
            ## Read the file with multiple space delimiter & Print the Sample data
            df = pd.read_csv(file_name, sep=r'\s+', engine='python')
            return "Success", df
        else:
            print("File not found.")
            df = pd.DataFrame()
            return "Error", df
        
    except requests.HTTPError as e:
        print("HTTP error occurred:", e)
        df = pd.DataFrame()
        return "Error", df
        

## Call the Download Function with File URL and File Name to be stored Locally
msg,df_carbon = download_file(data_url,file_name)
if msg == "Success":
    print("Sample Output from the File is below: \n")
    print(df_carbon.head())

    ## Define the number of Bins. We are using pd.cut to get the Bin Automatically created but we can do it manually as well
    num_bins = 8
    df_carbon['Bin'] = pd.cut(df_carbon['CO2'], bins=num_bins)

    ## Create frequency distribution
    frequency_distribution = df_carbon['Bin'].value_counts().sort_index()

    # Convert to DataFrame for easy use
    freq_df = frequency_distribution.reset_index()
    freq_df.columns = ['Range', 'Frequency']
    
    ## Display Frequency Distribution Output
    print("\n Frequency Distribution is as below: \n")
    print(freq_df)

    print("\n Histogram is as below \n")

    plt.figure(figsize=(8, 6))
    plt.bar(freq_df['Range'].astype(str), freq_df['Frequency'], color='skyblue', edgecolor='black')
    plt.title('Histogram from Frequency Distribution')
    plt.xlabel('CO₂ Emission Ranges (metric tons)')
    plt.ylabel('Number of Countries')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

else:
    print("Unable to Download the File")




```

**(b)Answer is as below**

```python
##  finding the mean, median, and standard deviation. Interpret each.

## Calculate statistics like Mean, Median & Standard Deviation
mean = df_carbon['CO2'].mean()
median = df_carbon['CO2'].median()
std_dev = df_carbon['CO2'].std()

## Display the Stats Calculated
print(f"Mean of CO2: {mean:.2f}")
print(f"Median of CO2: {median:.2f}")
print(f"Standard Deviation of CO2: {std_dev:.2f}")

print("\n\nInterpretation of the Statistics are as below:\n")

print("1. Mean 6.72 => Average CO2 emission by countries is 6.72. \n")
print("""
2. Median 5.90 => Median CO2 is 5.90 which is less than Mean 6.72.
It kind of highlights that it is right skewed and some countries with high CO2 is taking the mean high \n
    """)

print("""
3. Standard Deviation 3.36 => While Median is around 5.90, 3.36 is a kind of high standard deviation. 
It shows a high variability of CO2 among countries \n
    """)

```

## **Problem # 1.13.**
A report indicates that public school teacher’s annual salaries in New York city have an approximate  mean of `$`69,000 and standard deviation of `$`6,000. If the distribution has approximately a bell shape, report intervals that contain about (a) 68%, (b) 95%, (c) all or nearly all salaries. Would a salary of  `$`100,000 be unusual? Why? 


**Answer** Used Web Source to Read about the Bell Shape & Emphiral Rule: **https://www.geeksforgeeks.org/empirical-rule/**

```python
meanvalue = 69000
stddev = 6000

one_std_lowrange = meanvalue - stddev
one_std_highrange = meanvalue + stddev

two_std_lowrange = meanvalue - 2 * stddev
two_std_highrange = meanvalue + 2 * stddev

three_std_lowrange = meanvalue - 3 * stddev
three_std_highrange = meanvalue + 3 * stddev


print("\n** Report Interval 68% means that the 68% of the Salaries are in range of +- 1 Standard Deviation from Mean. \n")
print(f" > So the Salaries range would be between ${one_std_lowrange} and ${one_std_highrange} ")


print("\n** Report Interval 95% means that the 95% of the Salaries are in range of +- 2 Standard Deviation from Mean. \n")
print(f" > So the Salaries range would be between ${two_std_lowrange} and ${two_std_highrange} ")


print("\n** Report Interval nearly all means that the nearly of the Salaries are in range of +- 3 Standard Deviation from Mean. \n")
print(f" > So the Salaries range would be between ${three_std_lowrange} and ${three_std_highrange} ")

print(f"\n** Since $100,000 is more than ${three_std_highrange}, it is unusal Salary \n")
```

(a) Find the mean and standard deviation and interpret their values.

```python
import requests
import os
import pandas as pd
import matplotlib.pyplot as plt


## Specify the File URL & Local File Name
data_url = "http://stat4ds.rwth-aachen.de/data/Murder.dat"
file_name = "Murder.dat"

## Function to Download the File from the URL and Preview the Data
def download_file(file_url,file_name):
    try:
        response = requests.get(file_url)
        response.raise_for_status()

        ## Write to the File locally
        with open(file_name, "wb") as f:
            f.write(response.content)
        
        ## Verify if the file got downloaded successfully or not
        if file_name in os.listdir('.'):
            print("File downloaded successfully.\n")
            
            ## Read the file with multiple space delimiter & Print the Sample data
            df = pd.read_csv(file_name, sep=r'\s+', engine='python')
            return "Success", df
        else:
            print("File not found.")
            df = pd.DataFrame()
            return "Error", df
        
    except requests.HTTPError as e:
        print("HTTP error occurred:", e)
        df = pd.DataFrame()
        return "Error", df
        

## Call the Download Function with File URL and File Name to be stored Locally
msg,df_murder = download_file(data_url,file_name)
if msg == "Success":
    print("Sample Output from the File is below: \n")
    print(df_murder.head())

else:
    print("Unable to Download the File")
```

**a) Answer for Find the mean and standard deviation and interpret their values.**

```python
meanvalue = df_murder['murder'].mean()
stddevvalue = df_murder['murder'].std()

print(f"Mean is {meanvalue} and Standard Deviation is {stddevvalue}")

print(""" 
Since Standard Deviation is > 50% of the mean, it kind of shows a high variability among the values in the observation.
""")
```

**b) Answer for Find the five-number summary, and construct the corresponding box plot. Interpret.**

```python
import matplotlib.pyplot as plt



print(
    """
    Five Number Summary means: Minimum, Q1 - First Quartile (25%), Median (50%), Q3 - Third Quartile (75%) & Maximum value.\n
    We can derive this from Pandas Describe function
    """)

summary_df = df_murder.describe()

minvalue = summary_df.loc['min','murder']
q1value = summary_df.loc['25%','murder']
medianvalue = summary_df.loc['50%','murder']
q3value = summary_df.loc['75%','murder']
maxvalue = summary_df.loc['max','murder']
IQRvalue = q3value - q1value
LowerOutlier = q1value - 1.5*IQRvalue
UpperOutlier = q3value + 1.5*IQRvalue
rangeValue = maxvalue - minvalue

print("\n Five Number Summary is as below: \n")
print(f"1. Minimum = {minvalue} \n")
print(f"2. Q1 = {q1value} \n")
print(f"3. Median = {medianvalue} \n")
print(f"4. Q3 = {q3value} \n")
print(f"5. Maximum = {maxvalue} \n")

print("\n Some additional details for Outlier detection are as below: \n")
print(f"1. IQR = {IQRvalue} \n")
print(f"2. Lower Outlier Limit= {LowerOutlier} \n")
print(f"3. Upper Outlier Limit= {UpperOutlier} \n")
print(f"4. Range = {rangeValue} \n")

five_number_summary = summary_df.loc[['min', '25%', '50%', '75%', 'max'], 'murder']

## Creating a boxplot using matplotlib
fig, ax = plt.subplots()
box_data = [five_number_summary.values]
ax.boxplot(box_data, vert=False)

# Adding labels and title
ax.set_title(f'Boxplot for Murder - Five-Number Summary')
ax.set_yticklabels(['murder'])
ax.set_xlabel('Values')
plt.show()

print(
    f"""
    Interpretation as below: \n 
    1. Maximum 24.2 is the outlier since it is more than {UpperOutlier} \n
    2. Here Q1 is wider than Q3 meaning that there are few values that are higher than Median but concentrated
       that pulls the Median towards upper bound but many values are in lower bound spread out and not concentrated \n
    3. Data is Skewed
    """
)
```

```python
meanvalue = df_houses['price'].mean()
stdvalue = df_houses['price'].std()

minrange = meanvalue - stdvalue
maxrange = meanvalue + stdvalue

print(f"Mean: {meanvalue}")
print(f"Standard Deviation: {stdvalue}")

print(f"Now to find all the observation in the range {minrange} to {maxrange} ")

within_range = df_houses[(df_houses['price'] >= minrange) & (df_houses['price'] <= maxrange)]
count_within_range = within_range.shape[0] ## To get the number of rows

total_count = df_houses.shape[0]
percentage = (count_within_range / total_count) * 100

print(f"Count within ±1 std: {count_within_range} out of {total_count}")
print(f"Percentage: {percentage:.2f}%")

df_houses['price'].hist()

print(
    """
    As we can see from the Histogram, the Curve is Right Skewed and 68% as per Empirical Rule is for a Perfectly Normal Bell Curve

    In this case, since it is right skewed, this 68% didnt apply.
    
    """
)
```
