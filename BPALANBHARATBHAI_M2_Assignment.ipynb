## Problem # 2.4.

A wine connoisseur is asked to match five glasses of red wine with the bottles from which they
came, representing five different grape types.


(a) Set up a sample space for the five guesses.  
(b) With random guessing, find the probability of getting all five correct.

`(a) Your answer goes here`

**Details Provided:**

Given that there are 5 Glasses of Wine that is from 5 Bottles (with different graphe types). So Each Glass can be from any of 5 bottles.

To determine a single Glass from bottle, it would take max 5 permutations/combination. For 5 Glasses, it would take max 5 * 5 = 25 Permutations if we want to do it individually.

But if we need all 5 to be right at same time it would be 5! => 120 for matching 5 glasses to 5 bottles in a same event.

```python
from itertools import permutations

bottlelist = ['Bottle1','Bottle2','Bottle3','Bottle4','Bottle5']
glasslist = ['Wine1','Wine2','Wine3','Wine4','Wine5']

## Combination of Bottle + Glass List
all_combintations = []

for validate in permutations(bottlelist):
    pairing = list(zip(glasslist, validate))
    all_combintations.append(pairing)

print(f""" Total Number of Permutations possible are: {len(all_combintations)} \n """)

#Show the output of 5 items from the List
for i, p in enumerate(all_combintations[:5], 1):
    print(f"Pairing {i}: {p}")


```

`(b) Your answer goes here`

Now we need to find the probability of all 5 bottled guessed right out of 120 possible combinations. That is 1 of 120 outcome.

```python
all_right_prob = (1 / 120)

print(f""" Probability of all guess right is 1 of 120 => { all_right_prob:.4f} """)
```

## Problem # 2.17.

An instructor gives a course grade of B to students who have total score on exams and homeworks between 800 and 900, where the maximum possible is 1000. If the total scores have
approximately a normal distribution with mean 830 and standard deviation 50, about what proportion of the students receive a B?


`Your answer goes here`

**Details provided:**

1. Given that Student is graded B if the score is between 800 & 900.
2. Mean is 830 & Standard Deviation is 50

Based on research, we need to use Z-Score for calculating this.

```python
from scipy.stats import norm

mean_value = 830
std_dev = 50

start_score_stddev = (800 - mean_value)/std_dev
end_score_stddev = (900 - mean_value)/std_dev

# Use the standard normal CDF to compute probabilities
pro_lower = norm.cdf(start_score_stddev) 
pro_upper = norm.cdf(end_score_stddev) 

# Calculate the proportion between 800 and 900
b_grade_proportion = pro_upper - pro_lower

print(
    f"""
Proportion of students with B grades = {b_grade_proportion:.4f}
    """
     )


```

```python
## Download the File from the External URL using the code below

import requests
import os
import pandas as pd
import matplotlib.pyplot as plt


## Specify the File URL & Local File Name
data_url = "http://stat4ds.rwth-aachen.de/data/Income.dat"
file_name = "Income.dat"

## Function to Download the File from the URL and Preview the Data
def download_file(file_url,file_name):
    try:
        response = requests.get(file_url)
        response.raise_for_status()

        ## Write to the File locally
        with open(file_name, "wb") as f:
            f.write(response.content)
        
        ## Verify if the file got downloaded successfully or not
        if file_name in os.listdir('.'):
            print("File downloaded successfully.\n")
            
            ## Read the file with multiple space delimiter & Print the Sample data
            df = pd.read_csv(file_name, sep=r'\s+', engine='python')
            return "Success", df
        else:
            print("File not found.")
            df = pd.DataFrame()
            return "Error", df
        
    except requests.HTTPError as e:
        print("HTTP error occurred:", e)
        df = pd.DataFrame()
        return "Error", df
        

## Call the Download Function with File URL and File Name to be stored Locally
msg,df_income = download_file(data_url,file_name)
if msg == "Success":
    print("Sample Output from the File is below: \n")
    print(df_income.head())

else:
    print("Unable to Download the File")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the histogram with a density curve
plt.figure(figsize=(10, 6))

# Histogram (normalized to form a density)
sns.histplot(df_income, bins=30, kde=False, stat="density", color='skyblue', label="Histogram")

# Smooth density curve using kernel density estimation (KDE)
sns.kdeplot(df_income, color='darkblue', linewidth=2, label="KDE Curve")

# Labels and title
plt.xlabel("Income")
plt.ylabel("Density")
plt.title("Income Distribution: Histogram and Density Curve")
plt.legend()

plt.show()
```

## Problem # 2.21.

Plot the gamma distribution by fixing the shape parameter *k* = 3 and setting the scale parameter
= 0.5, 1, 2, 3, 4, 5. What is the effect of increasing the scale parameter? (See also Exercise 2.48.)

`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma


## Define different scale values
scales = [0.5, 1, 2, 3, 4, 5]

## Define x-axis values
x = np.linspace(0, 30, 500)

## Create plot
plt.figure(figsize=(10, 6))
for scale in scales:
    y = gamma.pdf(x, a=3, scale=scale)
    plt.plot(x, y, label=f'scale={scale}')

## Show plot
plt.title(f'Gamma Distribution (shape k={3})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Observation from Above Chart:**

We can see that as the value of Scale increases,

1. The peak decreases.
2. Peak is shifting towards the right - Mode Increases
3. Values are more spread - Variance increases.

## Problem # 2.27.

The distribution of *X* = heights *(cm)* of women in the U.K. is approximately *N*(162, $7^2$).
Conditional on *X = x*, suppose *Y* = weight *(kg)* has a *N*(3.0 + 0.40x, $8^2$) distribution. Simulate
and plot 1000 observations from this approximate bivariate normal distribution. Approximate
the marginal means and standard deviations for *X* and *Y*. Approximate and interpret the
correlation.


`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt

# Assign Values
mean_height = 162
stddev_height = 7
intercept = 3.0
slope = 0.40
stddev_weight_vs_height = 4

# Simulate X
X = np.random.normal(mean_height, stddev_height, 1000)

# Simulate Y given X
Y = np.random.normal(intercept + slope * X, stddev_weight_vs_height)

# Plot
plt.scatter(X, Y, alpha=0.5)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Simulated data of height and weight')
plt.grid(True)
plt.show()

# Marginal means and SDs
meanX = np.mean(X)
stdX = np.std(X)

meanY = np.mean(Y)
stdY = np.std(Y)

# Correlation
correlation = np.corrcoef(X, Y)[0, 1]

print(f"Mean of X (height): {meanX:.2f} cm")
print(f"SD of X (height): {stdX:.2f} cm")
print(f"Mean of Y (weight): {meanY:.2f} kg")
print(f"SD of Y (weight): {stdY:.2f} kg")
print(f"Correlation between X and Y: {correlation:.3f}")
```

**Based on the graph above, there is a positive linear co-relation between increase in height vs Weight**
